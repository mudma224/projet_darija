    This Jupyter notebook is dedicated to organizing and preparing our datasets.<br>We systematically import the files, remove unnecessary columns, and consolidate them into a single, unified dataframe for streamlined analysis.

```python
import os
import glob
import numpy as np 
import pandas as pd 
```

```python
#Import all Excel filenames into a list and map each one to its corresponding dataframe in the dictionary dfs.

filenames = glob.glob("BDD_darija/*.xlsx")

dfs = {}
for file in filenames: 
    name = os.path.splitext(os.path.basename(file))[0]
    dfs[name] = pd.read_excel(file)

dfs.keys()
```

```python
dfs['adjectives'].head()
```

```python
# Drop the columns n2, n3, n4, and eng from each dataframe
for name, df in dfs.items(): 
    df.drop(columns= ["n2", "n3","n4","eng"], inplace= True, errors= "ignore")

# Check how many dataframes contain at leat one of the specified columns

i=0
for name, df in dfs.items():
    existing = set(["n2", "n3", "n4", "eng"]) & set(df.columns)
    if existing:
        print(f"{name} has: {existing}")
        i += 1
print(f"{i} dataframes contain at leat one of the specified columns")
```

```python
# Check all dataframes with their colunms
for name, df in dfs.items(): 
    print(name, df.columns.tolist())
```

```python
# Merge all dataframes into one large dataframe

df_list = list(dfs.values())
data = pd.concat(df_list, ignore_index= True, axis= 0)
data.head()
```

```python
data.to_excel("data_fr.xlsx", index= False)
print("Save Successful")
```
