```python
import pandas as pd
import re
from deep_translator import GoogleTranslator
```

```python
from deep_translator import GoogleTranslator
files=['food.csv','clothes.csv','professions .csv']
translator = GoogleTranslator(source='en', target='fr')
dfs=[]
for f in files:
    temp = pd.read_csv(f, sep='\t', encoding='utf-16')
    temp = temp[['n1', 'eng']]
    temp['fren'] = temp['eng'].apply(lambda text: translator.translate(text) if isinstance(text, str) else text)
    temp = temp[['n1', 'fren']]
    dfs.append(temp)
df_final = pd.concat(dfs, ignore_index=True)
```

```python
df_final['fren'] = df_final['fren'].replace({
    'anneau':'bague',
    'capuchon':'casquette',
    'barbier':'coiffeur',
    'assistant':'secrétaire',
    'homme de volaille':'Volailler',
    'rendez-vous':'datte'
})
```

```python
def clean_text(text):
    if isinstance(text, str):
        # On supprime les caractères spéciaux avec re.sub
        text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
        return text.lower().strip()
    return text
df_final['n1']=df_final['n1'].str.lower().apply(clean_text)
df_final['fren']=df_final['fren'].str.lower().apply(clean_text)
df_final=df_final.dropna().drop_duplicates()
```

```python
df_final.to_csv("version_clean.csv", index=False, encoding='utf-8')
```
